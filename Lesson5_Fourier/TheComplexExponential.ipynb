```python
import numpy as np
import matplotlib.pyplot as plt
```

The imaginary number *i* in python is denoted 1j.

```python
print(np.power(1j, 2))
print(1j**2)
```

It can be used as the argument to a function.

```python
np.exp(1j)
```

```python
def cexp(t):
    return np.exp(-2*np.pi*1j*t)
```

```python
t1 = np.linspace(0,1,1000)
```

```python
plt.plot(t, cexp(t1).real)
plt.plot(t, cexp(t1).imag)
```

```python
plt.plot(cexp(t1).real, cexp(t1).imag)
plt.axis('equal')
```

```python
t2 = np.linspace(0, 0.3, 1000)
```

```python
plt.plot(cexp(t2).real, cexp(t2).imag)
plt.xlim(-1.2,1.2)
plt.ylim(-1.2,1.2)
plt.axis('equal')
```
